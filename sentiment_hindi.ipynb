```
# Merging all the text file data to form a single list which contains all text file statements
import os

Text_list = []                  # Creating empty list to store the text file data 

path_1 = '/content/drive/My Drive/emotions'         # path of the emotions folder
emotion = os.listdir(path_1)

for feel in emotion:
  print(feel) 
  path_2 = path_1 + '/' + feel
  folder = os.listdir(path_2)

  for file in folder:
    with open('/content/drive/My Drive/emotions/'+feel+'/'+ file, 'r') as f:
      rd = f.read()               # reading the data in the text file
      Text_list.append(rd)        # Adding the data to the list

print(Text_list)
```

```
print(len(Text_list))  # Total number of elements(Statements) in the list
```

```
# Selecting a random statement from the prepared list
import random

statement = random.choice(Text_list)
print(statement)
#--------------------------------------------------------#



# Translating the selected Hindi text to English using translate library
from translate import Translator

translator= Translator(from_lang="hindi", to_lang="english")
translation = translator.translate(statement)
print(translation, '\n')

#----------------------------------------------------------#



# Sentiment analysis of the English converted form of selected text using textblob
from textblob import TextBlob

obj = TextBlob(translation)
sentiment_polarity = obj.sentiment.polarity
sentiment_subjectivity = obj.sentiment.subjectivity

print("Polarity:", sentiment_polarity)
print("Subjectivity:", sentiment_subjectivity, '\n')

#----------------------------------------------------------------------------------#


if(sentiment_polarity == 0):
  print("Statement is Neutral")
elif(sentiment_polarity < 0):
  print("Statement is Negative")
else:
  print("Statement is Positive")
```
